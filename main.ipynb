```python
#opsional random
import random

g_r_awal = None
g_r_akhir = None
nilai = []
jml_data = None
def setacakdata(batas_random_awal, batas_random_akhir, jml_dataku):
  global g_r_awal, g_r_akhir, nilai, jml_data
  setnilai = []
  setnilai.append(batas_random_awal)
  setnilai.append(batas_random_akhir)
  randomin = [random.randint(batas_random_awal + 1, batas_random_akhir - 1) for _ in range(jml_dataku - 2)]
  setnilai.extend(randomin)
  setnilai.sort()

  g_r_awal = batas_random_awal
  g_r_akhir = batas_random_akhir
  nilai = setnilai
  jml_data = jml_dataku

setacakdata(50, 100, 1000)
```

```python
import math

def MinMaxKelasJangkauanInterval():
    global min_value, max_value, jumlah_kelas, jangkauan, interval
    min_value = min(nilai)
    max_value = max(nilai)
    jumlah_kelas = round(1 + 3.322 * math.log10(jml_data))
    jangkauan = max_value - min_value
    interval = math.ceil(jangkauan / jumlah_kelas)
    return  min_value, max_value, jumlah_kelas, jangkauan, interval

MinMaxKelasJangkauanInterval()
```

```python

def hitung_frekuensi(nilai, batas_bawah, batas_atas):
    count = 0
    for nilai_i in nilai:
        if batas_bawah <= nilai_i <= batas_atas:
            count += 1
    return count
```

```python
def getPandaFrame():   
    global df
    batas_bawah_kelas = min_value
    batas_atas_kelas = min_value + interval - 1
    kelas_list = []
    frekuensi_list = []
    median_list = []
    FiXi_list = []
    frekuensi_kumulatif_list = []

    # Menambahkan data ke list ketika perulangan berlangsung sebanyak kelasnya berdasarkan batas_atas_kelas
    for _ in range(jumlah_kelas):
        data_kelas = [nilai_i for nilai_i in nilai if batas_bawah_kelas <= nilai_i <= batas_atas_kelas]
        frekuensi = hitung_frekuensi(nilai, batas_bawah_kelas, batas_atas_kelas)
        if len(data_kelas) > 0:
            median_kelas = float((batas_bawah_kelas + batas_atas_kelas) // 2)
        FiXi = frekuensi * median_kelas
        kelas_list.append(f"{int(batas_bawah_kelas)}-{int(batas_atas_kelas)}")
        frekuensi_list.append(frekuensi)
        median_list.append(median_kelas)
        FiXi_list.append(FiXi)
        frekuensi_kumulatif_list.append(sum(frekuensi_list))
        batas_bawah_kelas = batas_atas_kelas + 1
        batas_atas_kelas += interval

    data = {
        "Kelas": kelas_list,
        "Frekuensi": frekuensi_list,
        "Median": median_list,
        "FiXi": FiXi_list,
        "Frekuensi Kumulatif": frekuensi_kumulatif_list
    }

    df = pd.DataFrame(data)
    return df

getPandaFrame()
```

```python
def getMean():
    mean = df['FiXi'].sum() / df['Frekuensi'].sum()
    return mean

getMean()
```

```python
def getMedian():
    total_frekuensi = df['Frekuensi Kumulatif'].iloc[-1]
    kelas_median_index = (df['Frekuensi Kumulatif'] >= total_frekuensi / 2).idxmax()

    tepi_bawah_kelas_median = (float(df['Kelas'].str.split('-').iloc[kelas_median_index][0]) + 
    float(df['Kelas'].str.split('-').iloc[kelas_median_index - 1][1]))/2

    total_frekuensi = df['Frekuensi Kumulatif'].iloc[-1]
    fik_sebelum_median = df['Frekuensi Kumulatif'].iloc[kelas_median_index - 1]
    fk_median = df['Frekuensi'].iloc[kelas_median_index]

    median = tepi_bawah_kelas_median + (((total_frekuensi + 1) / 2 - fik_sebelum_median) / fk_median) * interval

    return median, kelas_median_index, tepi_bawah_kelas_median, fk_median, fik_sebelum_median

getMedian()
```

```python
def getModus():
    kelas_modus_index = df['Frekuensi'].idxmax()

    tepi_bawah_kelas_modus = (float(df['Kelas'].str.split('-').iloc[kelas_modus_index][0]) + 
    float(df['Kelas'].str.split('-').iloc[kelas_modus_index -1][1]))/2

    fk_modus = df['Frekuensi'].iloc[kelas_modus_index]
    fk_sebelum_modus = df['Frekuensi'].iloc[kelas_modus_index - 1]
    fk_setelah_modus = df['Frekuensi'].iloc[kelas_modus_index + 1]

    b1 = fk_modus - fk_sebelum_modus
    b2 = fk_modus - fk_setelah_modus
    modus = tepi_bawah_kelas_modus + interval * (b1 / (b1 + b2))
    return modus

getModus()
```

```python
def hitung_desil(df, desil):
    desil_cari_hasil = desil / 10 * df['Frekuensi Kumulatif'].iloc[-1]
    desil_index = 0
    for index, row in df.iterrows():
      nilai_fk = row['Frekuensi Kumulatif']
      if desil_cari_hasil <= nilai_fk:
          break
      desil_index = desil_index + 1
    batas_bawah_desil = (float(df['Kelas'].str.split('-').iloc[desil_index][0]) 
    + float(df['Kelas'].str.split('-').iloc[desil_index - 1][1]))/2
    total_frekuensi = df['Frekuensi Kumulatif'].iloc[-1]
    fik_sebelum_desil = df['Frekuensi Kumulatif'].iloc[desil_index - 1]
    frekuensi_desil = df['Frekuensi'].iloc[desil_index]
    interval = float(df['Kelas'].str.split('-').iloc[0][1]) - float(df['Kelas'].str.split('-').iloc[0][0])

    hasil_desil = batas_bawah_desil + ((desil / 10 * total_frekuensi - fik_sebelum_desil) / frekuensi_desil) * interval
    return hasil_desil

hitung_desil(df, 9)
```

```python
def hitung_persentil(df, persentil):
    persentil_cari_hasil = persentil / 100 * df['Frekuensi Kumulatif'].iloc[-1]
    persentil_index = 0
    for index, row in df.iterrows():
      nilai_fk = row['Frekuensi Kumulatif']
      if persentil_cari_hasil <= nilai_fk:
          break
      persentil_index = persentil_index + 1

    batas_bawah_persentil = (float(df['Kelas'].str.split('-').iloc[persentil_index][0]) 
    + float(df['Kelas'].str.split('-').iloc[persentil_index - 1][1]))/2
    total_frekuensi = df['Frekuensi Kumulatif'].iloc[-1]
    fik_sebelum_persentil = df['Frekuensi Kumulatif'].iloc[persentil_index - 1]
    frekuensi_persentil = df['Frekuensi'].iloc[persentil_index]
    hasil_persentil = batas_bawah_persentil + ((persentil / 100 * total_frekuensi 
    - fik_sebelum_persentil) / frekuensi_persentil) * interval

    return hasil_persentil
hitung_persentil(df, 20)
```

```python
def setXiMinXbar():
    totalfixi = df['FiXi'].sum()
    global totalfrekuensi
    totalfrekuensi = df['Frekuensi'].sum()
    global xbar
    xbar = totalfixi / totalfrekuensi
    ximinxbar_list = []
    for index, row in df.iterrows():
        xi = row['Median']
        ximinxbar = xbar - xi
        ximinxbar_list.append(ximinxbar)
    df['xi-xbar'] = ximinxbar_list
    return df

setXiMinXbar()
```

```python
def setXiMinXbarKuadrat():
    ximinxbarkuadrat_list = []
    for index, row in df.iterrows():
        ximinxbar = row['xi-xbar']
        ximinxbarkuadrat = ximinxbar ** 2
        ximinxbarkuadrat_list.append(ximinxbarkuadrat)
    df['(xi-xbar)^2'] = ximinxbarkuadrat_list
    return df

setXiMinXbarKuadrat()
```

```python
def setFrekuensiXiMinXbarKuadrat():
    frekuensiximinxbarkuadrat_list = []
    for index, row in df.iterrows():
        frekuensiku = row['Frekuensi']
        ximinxbarkuadratnya = row['(xi-xbar)^2']
        frekuensiximinxbarkuadrat = frekuensiku * ximinxbarkuadratnya
        frekuensiximinxbarkuadrat_list.append(frekuensiximinxbarkuadrat)
    df['fi(xi-xbar)^2'] = frekuensiximinxbarkuadrat_list
    return df

setFrekuensiXiMinXbarKuadrat()
```

```python
def getFullTable():
    frekuensiximinxbar_list = []
    for index, row in df.iterrows():
        ximinxbar = row['xi-xbar']
        frekuensiku = row['Frekuensi']
        frekuensiximinxbar = abs(frekuensiku * ximinxbar)
        frekuensiximinxbar_list.append(frekuensiximinxbar)

    df['fi(xi-xbar)'] = frekuensiximinxbar_list
    return df
getFullTable()
```
